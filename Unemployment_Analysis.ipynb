```python
import pandas as pd
import numpy as np

df = pd.read_csv('Unemployment in India.csv')
df.columns = df.columns.str.strip()
print("Dataset Columns:", df.columns.tolist())
```

```python
print("missing values before cleaning data:\n", df.isnull().sum())
df=df.dropna()
df['Date']=pd.to_datetime(df['Date'].str.strip(),dayfirst=True)
df['Year']=df['Date'].dt.year
df['Month_Name']=df['Date'].dt.month_name()
print("\n Data Cleaning Complete")
print(f"Total rows remaining after cleaning:{len(df)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")
plt.rcParams['figure.figsize'] = (12, 6)
sns.lineplot(data=df, x='Date', y='Estimated Unemployment Rate (%)', color='green', linewidth=2.1)
plt.axvspan(pd.Timestamp('2020-03-25'), pd.Timestamp('2020-06-30'), color='red', alpha=0.2, label='Lockdown Shock Window')
plt.title('Unemployment Rate in Timeline: The COVID-19 Impact', fontsize=16, fontweight='bold')
plt.ylabel('Unemployment Rate(%)')
plt.xlabel('Date')
plt.legend()
plt.show()

```
